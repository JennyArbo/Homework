```python
# Don't modify this cell.
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# Don't modify this cell
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city in `cities` using a series of successive API calls.
* OPTIONAL: Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# OpenWeatherMap API Key
api_key = api_keys.api_keys

# base url for getting api data
base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key

# example request
#req = requests.get(base_url + f'&q={cities[0]}').json()
#print(req)

# Your code here. A loop maybe? 
#NOTE: API calls can be slow. They can also be limited. Once you have a list try not to rerun all API calls very often. 

city_weather=[]

#for city_ID in cities:
    #print(city_ID)
    #req = requests.get(base_url + f'&q={city_ID}').json()
    #city_weather.append(req)

#Doing this will print out the percent complete along with the name of the city
for i in range(len(cities)):
    print(i/len(cities)*100, cities[i])
    req = requests.get(base_url + f'&q={cities[i]}').json()
    city_weather.append(req)
    

    
```

### Convert Raw Data to DataFrame
* OPTIONAL: Export the city data into a .csv.
* Display the DataFrame

```python
#weather_df=pd.DataFrame(city_weather)
#weather_df.head(5)

#Want to return/store: City, Cloudiness, Country,Date, Humidity, Lat, Lng, Max Temp, Wind Speed

#Making a list
city_weather_tmp=[]
for entry in city_weather:
    if entry["cod"]=="404":
        continue
    city_weather_dict={}
    city_weather_dict["City"]=(entry["name"])
    city_weather_dict["Cloudiness"]=entry["clouds"]["all"]
    city_weather_dict["Country"]=entry["sys"]["country"]
    city_weather_dict["Date"]=entry["dt"]
    city_weather_dict["Humidity"]=entry["main"]["humidity"]
    city_weather_dict["Lat"]=entry["coord"]["lat"]
    city_weather_dict["Lng"]=entry["coord"]["lon"]
    city_weather_dict["Max Temp"]=entry["main"]["temp_max"]
    city_weather_dict["Wind Speed"]=entry["wind"]["speed"]
    city_weather_tmp.append(city_weather_dict)
    
weather_df=pd.DataFrame(city_weather_tmp)
weather_df.head()
print(len(weather_df))


#Saving the data to a csv file
weather_df.to_csv(r'/Users/jenniferarbuszewski/documents/JennyArboCoderGirl/Homework/Individual_Homework/week_05/CurrentWeather_OpenWeatherMap.csv', index = False)
   
```

```python
# OpenWeatherMap API Key
api_key = api_keys.api_keys

# base url for getting api data
base_url_p = "http://api.openweathermap.org/data/2.5/air_pollution?&APPID=" + api_key


# example request
#req = requests.get(base_url + f'&q={cities[0]}').json()
#print(req)

# Your code here. A loop maybe? 
#NOTE: API calls can be slow. They can also be limited. Once you have a list try not to rerun all API calls very often. 

coord_pollution=[]

#for city_ID in cities:
    #print(city_ID)
    #req = requests.get(base_url + f'&q={city_ID}').json()
    #city_weather.append(req)

#Doing this will print out the percent complete along with the name of the city
for j in range(500):
    print(j/500*100, coords[j])
    lat_p=coords[j][0]
    lon_p=coords[j][1]
    req_p = requests.get(base_url_p + f'&lat={lat_p}&lon={lon_p}').json()
    coord_pollution.append(req_p)

    

    
```

```python
print(coord_pollution)
```

```python
#pollution_df=pd.DataFrame(coord_pollution)
#pollution_df.head(5)

#Want to return/store: City, Cloudiness, Country,Date, Humidity, Lat, Lng, Max Temp, Wind Speed

#Making a list
coord_pollution_tmp=[]
for place in coord_pollution:
    #if place["cod"]=="404":
      #  continue
    coord_pollution_dict={}
    coord_pollution_dict["Lat"]=place["coord"]["lat"]
    coord_pollution_dict["Long"]=place["coord"]["lon"]
    coord_pollution_dict["AQI"]=place["list"][0]["main"]["aqi"]
    coord_pollution_dict["CO"]=place["list"][0]["components"]["co"]
    coord_pollution_dict["NO"]=place["list"][0]["components"]["no"]
    coord_pollution_dict["NO2"]=place["list"][0]["components"]["no2"]
    coord_pollution_dict["O3"]=place["list"][0]["components"]["o3"]
    coord_pollution_dict["SO2"]=place["list"][0]["components"]["so2"]
    coord_pollution_dict["Fine Particulate Matter"]=place["list"][0]["components"]["pm2_5"]
    coord_pollution_dict["Coarse Particulate Matter"]=place["list"][0]["components"]["pm10"]
    coord_pollution_dict["NH3"]=place["list"][0]["components"]["nh3"]
    coord_pollution_tmp.append(coord_pollution_dict)
   

    
pollution_df=pd.DataFrame(coord_pollution_tmp)

print(len(pollution_df))
pollution_df.head()
```
