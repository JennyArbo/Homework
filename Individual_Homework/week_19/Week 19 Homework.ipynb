### Linear Regression (univariate or multivariate)

**Summary:** Linear regression is a supervised learning model that can be used for continuous variables. It may be perfomed using either univariate or multivariate features. For a univariate model, the regression finds the relationship between a target (dependent) variable and a single governing (independent) variable. For a multivariate model, the relationship between a target (dependent) variable and multiple governing (independent) variables is explored. In a linear regression, a line relating the variables is drawn. This is known as a best fit line when the errors are optimized/minimized. There are a variety of ways in which to identify a best fit line, but all rely on minimizing the distance between the predicted dependent variable and the true/known dependent variable values. The difference between predicted and true value is known as the error and may also be described as a residual. (Of note, sometimes exploring the residual errors can lead to interesting insights about both the data and the model. For example, the residuals may themselves reveal a pattern that can help illuminate additional factors that may be influencing the dependent variable or may reveal something about ways to improve the model construction itself). 

**Evaluation:**
As mentioned in the previous section, the error between the predicted and true/known dependent variable should be minimized in a linear regression problem. There are various ways to calculate and examine those errors. $R^{2}$ (aka the coefficient of determination) , mean squared error (MSE) and root mean squared error (RMSE) or commonly used to evaluate the performance of a linear regression model. $R^{2}$ varies between 0-1, where 1 indicates a perfect fit of the model to the data (a value of 1 is exceedingly rare and if achieved, it should be regarded with skepticism and be evaluated further). It is calculated by squaring the correlation coefficient (r). The $R^{2}$ value is generally regarded as a measure of the goodness of fit of the model to the data. The mean squared error is represented by the following formula: MSE = 1/n * $\Sigma$ ($y_{i}$ - $\hat{y}_{i}$)$^{2}$ where i varies from 1 to n, n = # of samples, and $\hat{y}$ represents the predicted values and y represents the actual/known values. Because MSE is often difficult to interpret, RMSE may often be used instead. RMSE is calculated by taking the square root of the MSE, making the results more interpretable as they are in the same units as the dependent variable. Unlike $R^{2}$, the MSE or RMSE ideally is as low as possible.

### Logistic Regression (univariate or multivariate)

**Summary:**
Logistic regression techniques may work well for single of multi-class categorical data. This is a supervised learning model that aims to assign the dependent (target) variable to a known class. The logistic regression is essentially a probability problem, where the user can determine a cutoff of threshold value (called a decision boundary) for when one condition is more likely than another (for example, a plant is an oak based on say leaf shape). The classification can be binary (for example, a tumor is cancerous or not), or multinomial (without ordering: say an insect could be a butterfly, a ladybug, or a grasshopper or with ordering: restaurant reviews with a 1-5 star rating). 

**Evaluation:**
Unlike the linear regression, $R^{2}$, MSE, and RMSE are not applicable for evaluation of a logistic regression. Rather, precision, accuracy, F1 score, area under the curve (AUC) and/or recall are commonly used for evaluation. A confusion matrix can be an especially powerful evaluation tool. In some classification problems, the false positive or false negative rate may lead to significant negative consequences. The confusion matrix can help visualize and assess these false positive and negative rates. I will summarize each of these measures below:

Precision:
Precision yields information about how many predictions are correctly classified as positive compared to the total number of positive predictions made. This quantity is given by the following formula:
True Positives / (True + False Positives)

Recall:
Recall yields information about how the number of predicted positives compares to the number of true positives and false negatives. 
True Positives / (True Positives + False Negatives)

Accuracy Score:
(True Positives + True Negatives) / (True Negatives + True Positives + False Negatives + False Positives)
The accuracy score indicates how many samples were correctly predicted. The concern with accuracy scores is that they are sensitive to unbalanced data and may change when the make up of the data presented to the model evolves AND has a different balance of classes compared to the training data. It also does not give any information about the false negatives vs. false positive cases, which can be vital in some applications. So, at times, accuracy score alone may be misleading in terms of understanding model performance.

F1 Score:
The F1 score combines information about both precision and recall to give a more comprehensive look at model performance. It is given by the formula:
(2 * Precision * Recall) / (Precision + Recall)
A higher F1 score is indicative of a more performant model. 


Confusion Matrix: The confusion matrix provides a simple visual way to evaluate a model. It shows predicted vs actual positive and negative results. The result is a graphical representation of true positive/negative values and false positive/negative values (where the model got "confused").

The KNN algorithm works by assuming that datapoints that are close together in feature space most likely have similar characteristics. ### Decision Tree

**Summary:**
Like KNN, decision trees may be applied to both classification and regression problems. Decisions trees tend to be more transparent and easy to visualize the decision making process than other algorithms. Decision trees are composed of nodes, leaves, and branches. Nodes = features or classes, branches = decision,  and leaves = outcomes. 

Decision trees operate by asking a series of questions before coming to a final answer/categorization. For a very short example, we could start off with a question like  "does it have feathers?". If yes, we could then ask if it has wings? If yes, maybe we would now return the answer of "it is a bird". Essentially, the decision tree keeps breaking down a large dataset into smaller segments until such a time as we can assign a value or class designation to our target variable. In an ideal case, the data would keep being split until there is a clear answer (in other words, each subset is "pure"). Splitting of the data in the decision tree algorithm is selected for optimal information gain at each step. When the data no longer needs to be split, we have arrived at a leaf aka a final outcome. Decision trees are notably good at dealing with data that may have a high noise:sample ratio and datasets with missing data (sparse data). Drawbacks of this model type are that it can be unstable and prone to overfitting, however there are techniques to minimize those problematic outcomes (for example, bagging and boosting may help address overfitting).

**Evaluation:**
 The method of evaluation will depend upon whether we are using decision tree for a regression or a classifcation problem. For a regression, RMSE or MSE would likely be a good choice to evaluate performance. For classification, use of the accuracy, precision, F1 score, and/or visually with a confusion matrix (see above answers for more details).

### Random Forest

**Summary:**
Random forests may be used for either regression or classification problems. Unlike the previous techniques discussed so far, a random forest is known as an ensemble method which relies upon the use of combining multiple decision trees. Random forests are built on the idea of bagging. In bagging, bootstrapping (random sampling with replacement) is used to create a number of decision trees from the given training data. The goal of using bagging is to decrease the variance without simultaneously increasing the bias. The advantage of this is the model is given more chances to run on slightly different sets of data without having to collection many times the original amount of data. Each tree has a tendency to be overfitted due to the algorithm construction; however, this tendency may actually be viewed as a positive as it accentuates differences between the various trees. Model performance is improved by the inclusion of multiple trees and the exposure of the models to slightly different training sets. The method of determining the final outcome varies for the regression and classification problems. In regression, the final outcome for the target variable is expressed as the average of the outcomes from each decision tree in the ensemble. For the classification problem, the outcomes are pooled and then the final determination is decided by majority vote, where the class that appears most frequently is then assigned. It is also important to note that every time the model is run, slightly different results will be obtained. This is not due to instability but is actually built into the algorithm and is the result of the inclusion of randomness each and every time the model is run (even with a set random state).

**Evaluation:**
A random forest model uses the same evaluation metrics as the other models discussed so far. For classification problems, precision, accuracy, recall, etc may be appropriate. For regression, RMSE/MSE would be potentially good metrics.

**Other Notes:**
Number of random features included at split points is a very important parameter to consider and evaluate (one suggestion I read was to try using 1/3 the number of input variables for regression and the square root of the number of input variables for classification). ### XGBoost

**Summary:**
Like random forest, the XGBoost algorithm builds upon the decision tree model and is an ensemble method. It has gained popularity recently and is now a frequently used machine learning algorithm. XGBoost stands for "extreme gradient boosting". Where the random forest uses the concept of bagging, XGBoost, as implied by its name, relies on bagging. The algorithm is popular both due to its computation speed and also, oftentimes, the enhanced model performance. It can be applied to both regression and classification problems.

Let's break apart the different components of the XGBoost algorithm. First, let's consider what boosting does and why it can lead to enhanced model performance. Whereas the random forest contains a multitude of individual decision trees, in boosting, the models are run in serial instead of in parallel. In effect, the outcome of one model is fed into the next model and so on, in order to increase the predictive power. The models are somewhat weak to begin and with each new model run, some weaknesses/errors are dealt with. As a result, with each iteration, the model can be improved slightly and by the time the final output is given, there should be increased model efficacy. The idea there is to select for the highest performing models, thus "boosting" them and giving the technique its name. Like all models, the main idea is to minimize the loss function (in essence, the difference between predicted and known/true values for the target variable). The chosen loss function depends on the question asked and the types of data available (a common example for a regression question would be the use of least squares). The gradient part of the name comes from using a differential loss function.

There are other gradient boosting methods available but XGBoost does have some unique features that in some cases make it more desirable. In particular, it is able to handle missing data and is optimized to minimize overfitting.

This algorithm generally produces interpretable results and is typically more performant than either a single decision tree or the random forest models. It can also help with feature selection and is not especially prone to overfitting. It generally performs well for datasets that are not super large (extremely large datasets are the domain of neural networks, in general). However, due to the sequential nature of gradient boosting algorithms, it can be difficult to scale up and the training process is necessarily longer. The XGBoost algorithm gets around this long training time and it optimized to the point that training may actually take even less time than simpler models. It also tends to be somewhat sensitive to outliers.

### Neural Networks

**Summary:**
Neural networks are a machine learning concept roughly built on the idea of processing of data by the human brain,  and in particular, on the transfer of data by neurons. The basic structure of a neural network includes an input layer (the initial data), a number of hidden layers (determined when the network is built and where data is passed along a system of nodes) and an output layer (the final "answer" or results). The hidden layers are comprised of nodes. Each node is connected to a number of nodes in the next deepest layer. When the number of layers exceeds 2, this is considered deep learning. These nodes are the stand ins for neurons and function similarly (to a neuron) at a surface level. Each node essentially performs it's own modeling (often similar to a linear regression). If certain conditions are met in a given node, the node is "activated" and passes along information/data to another node where the process is repeated. If not, the node does nothing. This process continues through the given number of hidden layers with the data being refined and transformed along the way. The final result is the output layer which gives the "answer" to the initial situation set up when the network is initialized. 

**General steps to build a neural network:**
1. Clean and preprocess data
2. Input data and split into training and test sets
3. Select an activation function (may be useful to try a few to optimize the network response)
4. Train the network 
5. Make tweaks to learning rate, weights, etc (can use backward propagation) to tune the model
6. Test your model. If acceptable, move forward and if not, try tuning and adjusting parameters. If still unacceptable (hopefully if you've done enough and high quality preprocessing this won't happen), think and make sure that data was high enough quality and plentiful enough, that the question suits the data, etc.

**Evaluation:**
The exact metrics chosen to evaluate a neural network will depend on the type of data, research question, field, etc. In general, however, the performance of a neural network will be determined by some type of loss function. The specifics will vary dependant on whether the question at hand is a regression, classification, or ranking problem.

Examples of loss functions that could be used to evaluate neural network performance (not at all exhaustive) could include: 

for regression: mean absolute error (L1), mean squared error (L2),

for classification: negative log likelihood (NLL), cross entropy, or hinge embedding, or Kullback-Liebler divergence (multiclass classification),

for ranking: margin ranking 

or for content based retrieval, triplet margin

Any loss functions works by evaluating the mismatch between known/true values and the predicted model output. Basically, they consider how far off the model is from what we expect as output. In any case, loss should be minimized but the way we do that is highly dependent on the model and research question. Some loss functions respond more to outlier values than others do, some punish large mistakes more than small ones, etc. Know your data and your model and choose an appropriate loss function accordingly.

### Clustering (K-Means is specifically described here)

**Summary:**
Simply put, the k-means clustering algorithm looks for patterns in a dataset and based on the available information, assigns the datapoints to k different clusters. Datapoints that share certain characteristics are considered to be similar and will be assigned to the same cluster. The value for k must be determined to find the optimal number of clusters. The ultimate goal is for each data point to belong to a single cluster. The algorithm compares the sample value to the centroid (the mean value for the cluster) for each cluster and considers the distance between the centroid and the data points and tries to minimize that distance. This may be done using either inertia or sum of square distances. 

The first step is to randomly assign centroid values for the specified k number of clusters. The algorithm then iterates over centroid values until the centroid values are no longer changing (i.e. stable). Once the clusters are determined with optimal centroids, the distances between data points and centroids is calculated; the data point is assigned to the cluster whose centroid is closest to it. K-means tends to do well with data with simple geometric structure and performs poorly with more complicated structures. More weight also tends to be placed on larger clusters, which can be problematic. This method also works poorly for data points that are not geometrically close to one another but should nonetheless be placed in the same cluster. Visualization is also difficult for high dimension problems. Applying dimensionality reduction techniques prior to running the model was result in enhanced performance and decreased computational expense.

**Determination of Number of clusters & Model Evaluation:**

*Elbow method:* 
The elbow methods gives an evaluation criteria for determining the optimal number of clusters. It works by computer the sum of squared distances between each data point and the centroid value for each specified value of k. The distances can then be plotted with k value to find the optimal distance. A smaller number of clusters is usually better so there is some degree of personal opinion on where the cutoff should be made. This method may also not work particularly well for determining cluster number in k-means analysis.

*Silhouette analysis:* 
This method is a means of assessing the separation of the different clusters (in general, more separation is better). Coefficient values vary from -1 to 1. A 0 value indicates low separation, 1 indicates high degree of separation, and -1 indicates that the sample has been assigned incorrectly to a cluster. 

Silhouette ane elbow methods may be compared to yield more insight. As far as I can tell though, there is no absolutely definitive answer on how to determine optimal k value.

### Hierarchical

**Summary:**
Hierarchical clusters is an unsupervised learning method that makes use of a tree structure called a dendogram. These methods can either take a top down or bottom up approach to assigning clustering. In top down, all of the data is initially assigned to a single cluster which is then successively broken apart into smaller clusters. In bottom up, each point is assigned its own cluster then the clusters are then successively merged together. 

In agglomerative clustering, each individual datapoint is assigned its own cluster. The algorithm then merges clusters together based on similar features. There are several ways in which the clusters may be merged; for example the "ward" approach computes the sum of squares and finds an optimal value for each cluster. A drawback to agglomerative clustering is it can be very computationally expensive. Constraints about how clusters can be merged and rules as to which can be combined can be put in place. Connectivity matrices can be implemented to add in these various constraints. The division method is almost the opposite of the agglomerative method. It is a top down approach and splits an initial single cluster into multiple smaller clusters. 

There are various methods of determining cluster separation (the distance between clusters). These methods are known as linkages. In general, there are 5 types of linkages (single, complete, average, centroid, and ward's). 


**Determination of Number of Clusters & Evaluation:**
The determination of cluster number can be determined using either the dendogram or a bar chart. The largest height difference in the dendogram represents where the threshold for number of clusters should be made. A barplot can be similarly used by plotting the height from the dendogram for each number of clusters.

### Dimensionality Reduction

**Summary:**

Dimensionality reduction is often applied to a dataset as part of preprocessing. There are a number of benefits, including leading to less complex models and removal of redundant and multicolinear data which can lead to improved algorithm performance. The main goal of any type of dimensionality reduction is to minimize the number of features/dimensions of feature space.There are a number of types of dimensionality reduction but a popular one is Principle components analysis (PCA). In PCA, multicolinear features are combined linearly such that the principle components are no longer correlated. The first few PCs are responsible for the largest amount of sample variance. Therefore, it is possible to take just a few PCs to represent the data in a different model, significantly reducing (potentially) the number of dimensions needed to describe the dataset. 

**Evaluation:**

The evaluation is not as straight forward as for some other methods. Looking at final model performance with different numbers of PCs, different numbers of features from other types of feature selection, etc can give some indications about if appropriate choices have been made. For PCA a threshold can also be identified for how much variance you would like explained and the number of PCs can be selected accordingly.

### 4) Perform k-means on this dataset for customer segments. Customer segments help determine what types of people buy your product, which allows you to target more people like your usual customers. Should you look at all the data, or which subset of data should you use? What is the ideal number of clusters? Which approach did you use to find the ideal number of clusters and why?


It is apparent that the dataset is highly imbalanced when we look at Country. The vast majority of customers are from the United Kingdom. In terms of identifying the type of person who buys these products, we can say that the largest share (by a huge margin) all come from the same country. As a result, it is reasonable to focus the analysis moving forward on customers and transactions just from the United Kingdom. 

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

inertias = []
distortions = []
for k in range(1,10):
    #build and fit the model
    model = KMeans(n_clusters=k, init = 'k-means++').fit(scaled)
    
    #average of squared distance from cluster center to the data points
    distortion = sum(np.min(cdist(scaled, model.cluster_centers_,
                                  'euclidean'),axis =1))/scaled.shape[0]
    
    distortions.append(distortion)
    
    inertia = model.inertia_
    inertias.append(inertia)
    
    
    
#From Mounika
#from sklearn.cluster import KMeans
#wcss = []
#for k in range(1,11):
   # kmeans = KMeans(n_clusters=k, init="k-means++")
   # kmeans.fit(grouped_customerid.iloc[:,1:])
   # wcss.append(kmeans.inertia_)
#plt.figure(figsize=(12,6))    
#plt.grid()
#plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
#plt.xlabel("K Value")
#plt.xticks(np.arange(1,11,1))
#plt.ylabel("WCSS")
#plt.show()
```

```python
plt.plot(range(1,10),distortions)
plt.title('Elbow Method with Distortion')
```

```python
plt.plot(range(1,10),inertias)
plt.title('Elbow Method with Inertia')
```

**Ideal Number of Clusters:**

Using the elbow method with inertria, it appears that 7 clusters would be optimal. For the elbow method using distortion, one could argue 3 or maybe 6/7 clusters would be ideal. Since the inertia method didn't really have a clear inflection point, it is more difficult for me to interpret.
